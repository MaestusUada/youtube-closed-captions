```
from __future__ import unicode_literals

import youtube_dl


class MyLogger(object):
    def debug(self, msg):
        pass

    def warning(self, msg):
        pass

    def error(self, msg):
        print(msg)


def my_hook(d):
    print(d)
    if d['status'] == 'finished':
        print('Done downloading, now converting ...')


ydl_opts = {
    'writesubtitles': True,
    'subtitlesformat': 'vtt',
    'subtitleslang': 'tr',
    'playlistrandom': True,
    # 'listformats': True,
    'logger': MyLogger(),
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
        'preferredquality': '192',
    }],
    'sleep_interval': 5 ,
    'progress_hooks': [my_hook],
}

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    meta = ydl.extract_info("https://www.youtube.com/playlist?list=PL_VIYA-L9VnKTrFDAWpAUlm9FcVNOxn02")
```
